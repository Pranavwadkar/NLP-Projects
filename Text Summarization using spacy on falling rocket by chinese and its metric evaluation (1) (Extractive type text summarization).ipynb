```python
text ="""Experts have predicted that the latest piece of big Chinese space junk will fall back to Earth but are not sure where exactly it will land. The re-entry will be around the end of the month, probably on July 31 at 02:22 UTC ± 17 hours. The mentioned rocket is roughly 21-tonne and was part of the Wentian space station module. It was launched on Sunday and docked successfully with the Tiangong space station. A Long March 5B (CZ-5B) rocket propelled the uncrewed Wentian spacecraft from the Wenchang launch centre on China’s tropical island of Hainan. It appeared to be concerning as experts have claimed that they are not sure where the debris will land. Holger Krag, the head of the Space Safety Program Office for the European Space Agency, told SpaceNews: “It is always difficult to assess the amount of surviving mass and number of fragments without knowing the design of the object, but a reasonable ‘rule-of-thumb’ is about 20-40 per cent of the original dry mass.”"""
```

```python
print(text)
print(len(text))
```

```python
## Text cleaning

for i in text:
    if (i=="-" or i =="." or i=="," or i=="'" or i=="-"):
        text=text.replace(i," ")    
print(text)
```

```python
## Removing stop word

import nltk
from nltk.corpus import stopwords as STOP_WORDS

tokenised_words = nltk.word_tokenize(text)

count =0
for i in tokenised_words:
    if i in STOP_WORDS.words('english'):
        tokenised_words.remove(i)
        count+=1

print(tokenised_words)
```

```python
## Word frequency table

count =0
dict1={}

##First we have to assign each of the tokens as the key pair in the dictionary


for i in tokenised_words:
    count+=1
    dict1[i]=0
    

## Now for creating the value of the word
word_count=0

for i in tokenised_words:
    word_count=0
    for j in tokenised_words:
        if(i == j):
            word_count+=1
            dict1[i]=word_count

print(dict1)
```

```python
### Performing the clustering
import spacy
from string import punctuation
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(text)
```

```python
print(type(doc))
```

```python
## making the list of the tokens 

token= [token.text for token in doc]
print(token)
```

```python
max_frequency = max(dict1.values())
print(max_frequency)
```

```python
for i in dict1.keys():
    dict1[i]=dict1[i]/max_frequency
    
    print(dict1)
    
```

```python
## Sentence tokensisation

sentence_tokens = [sent for sent in doc.sents]
print (sentence_tokens)
```

```python
sentence_scores = {}
for sent in sentence_tokens:
    for word in sent:
        if word.text.lower() in dict1.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent] = dict1[word.text.lower()]
            else:
                sentence_scores[sent] += dict1[word.text.lower()]
```

```python
sentence_scores
```

```python
from heapq import nlargest
```

```python
select_length = int(len(sentence_tokens)*3)
print(select_length)
```

```python
summary = nlargest(select_length,sentence_scores,key = sentence_scores.get)
final_summary = str(summary[0])
print(final_summary)
print(len(final_summary))
```
